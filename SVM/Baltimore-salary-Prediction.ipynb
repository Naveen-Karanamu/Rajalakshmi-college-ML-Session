# Feature Scaling

* Variablizing the data

```python
x=np.array(newdff.drop('annualsalary',axis=1))
y=np.array(newdff.annualsalary)
```

* Creating StandardScalar variable

```python
stanScale=StandardScaler()
```

* Transforming the Data into the StandardScalar

```python
x=stanScale.fit_transform(x.reshape(-1,1))
y=stanScale.fit_transform(y.reshape(-1,1))
```

* Splitting the Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Without Feature Scaling

```python
a=np.array(newdff.drop('annualsalary',axis=1))
b=np.array(newdff.annualsalary)
```

* Splitting the Data

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2)
```

* Selecting the model

```python
model=SVR()
```

* Fitting the model

```python
model.fit(a_train,b_train)
```

# Performance and Errors

* Checking the Accuracy

```python
model.score(a_test,b_test)
```

* Prediction

```python
b_pred=model.predict(a_test)
```

* R squared Test

```python
r2_score(b_test,b_pred)
```

* MSE

```python
mean_squared_error(b_test,b_pred)
```

* MAE

```python
mean_absolute_error(b_test,b_pred)
```

* Plotted the Actual Marks v/s Predicted Marks Graph

```python
plt.scatter(b_test,b_pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual Marks v/s Predicted Marks")
plt.grid()
```
